```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
def TML(file, file_name):

    p_response = []

    for n in range(10,21):
        if n != 15:
            n_higher = 0
            for k, row in file.iterrows():
                if row[2] == n:
                    if row[6] == 'before':
                        n_higher += 1
                if row[4] == n:
                    if row[6] == 'after':
                        n_higher += 1
        
            p_response.append(n_higher/10)
    
    x = [10,11,12,13,14,16,17,18,19,20]
    y = p_response
    plt.scatter(x,y)
    plt.title(file_name, fontsize=20)
    plt.xlabel('Frequency(Hz)', fontsize=15)
    plt.ylabel("proportion 'higher' responses", fontsize=15)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.ylim(-0.1,1.1)
    plt.grid()
```

```python
TML_list = next(os.walk('./TM'))[2]
plt.figure(figsize=(30,60))

for i,j in enumerate(TML_list):
    TML_file = pd.read_csv('./TM/{}'.format(j), sep='\t', header=None)
    plt.subplot(8,3,i+1)
    TML(TML_file, j[:5])
```
