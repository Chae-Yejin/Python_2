Sort 528 trials in Test for 8 different response types and calculate the probability for each of them

```python
import pandas as pd
import numpy as np
import os
```

```python
def memory_scores(subject_ID):
    type_0 = 0
    type_1 = 0

    for j, row in test_file.iterrows():
        if row['Task_Type'] == 0:
            type_0 += 1
            if row['Res1'] == 1:
                p_test.loc[subject_ID, 'HIT-HC'] += 1
            elif row['Res1'] == 2:
                p_test.loc[subject_ID, 'HIT-LC'] += 1
            elif row['Res1'] == 3:
                p_test.loc[subject_ID, 'MISS-LC'] += 1
            else:
                p_test.loc[subject_ID, 'MISS-HC'] += 1
        else:
            type_1 += 1
            if row['Res1'] == 1:
                p_test.loc[subject_ID, 'FA-HC'] += 1
            elif row['Res1'] == 2:
                p_test.loc[subject_ID, 'FA-LC'] += 1
            elif row['Res1'] == 3:
                p_test.loc[subject_ID, 'CR-LC'] += 1
            else:
                p_test.loc[subject_ID, 'CR-HC'] += 1
    
    for k in p_test.columns[0:4]:
        p_test.loc[subject_ID, k] /= type_0
        
    for n in p_test.columns[4:]:
        p_test.loc[subject_ID, n] /= type_1
```

```python
dir_list = next(os.walk('./DMdata'))[1]


p_test = pd.DataFrame(index=dir_list, 
                      columns=['HIT-HC', 'HIT-LC', 'MISS-LC', 'MISS-HC',
                              'FA-HC', 'FA-LC', 'CR-LC', 'CR-HC'])
p_test.fillna(0, inplace=True)


for i in dir_list:
    test_file = pd.read_csv('./DMdata/{}/{}_test.txt'.format(i,i), sep='\t', index_col=False)
    memory_scores(i)
```
